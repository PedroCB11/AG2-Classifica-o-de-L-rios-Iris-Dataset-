```python
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)
IQR = q3 - q1

limite_inferior = q1 - 1.5 * IQR
limite_superior = q3 + 1.5 * IQR

outliers = df[(df['age'] < limite_inferior) | (df['age'] > limite_superior)]

len(outliers)
```

## <font color='green'>**2.6 Correlação Simples**</font>

### <font color='cyan'> Q18: Escolha duas características numéricas do conjuntos de dados. Escreva seus nomes a seguir.</font>

age e capital-gain

### <font color='cyan'> Q19: Calcule e exiba o Coeficiente de Correlação de Pearson ($r$) entre elas.</font>

```python
df[['age', 'capital-gain']].corr(method='pearson')
```

### <font color='cyan'> Q20: Interprete o coeficiente de correlação, em termos de sinal e magnitude.</font>

```python
r = df['age'].corr(df['capital-gain'])

print("Coeficiente de Correlação de Pearson (r):", r)

# interpretação do sinal
if r > 0:
    sinal = "positiva"
elif r < 0:
    sinal = "negativa"
else:
    sinal = "nula"

# interpretação da magnitude
abs_r = abs(r)

if abs_r < 0.2:
    magnitude = "muito fraca"
elif abs_r < 0.4:
    magnitude = "fraca"
elif abs_r < 0.6:
    magnitude = "moderada"
elif abs_r < 0.8:
    magnitude = "forte"
else:
    magnitude = "muito forte"

print(f"A correlação é {sinal} e de magnitude {magnitude}.")
```
